```python
from __future__ import absolute_import, division, print_function, unicode_literals

import os
import pickle
import time

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv('1.csv',names=['date', 'freq'])
df = df.dropna(axis=0,how='any')
df['date'] = pd.to_datetime(df['date'])
# df.reset_index(drop=True,inplace=True)
df = df.set_index('date')
df.head(5)
# print(len(df))
```

```python
df.plot()
plt.show()
```

```python
data_FT = df['freq']
```

```python
close_fft = np.fft.fft(np.asarray(data_FT.tolist()))
fft_df = pd.DataFrame({'fft':close_fft})
fft_df['absolute'] = fft_df['fft'].apply(lambda x: np.abs(x))
fft_df['angle'] = fft_df['fft'].apply(lambda x: np.angle(x))
```

```python
plt.figure(figsize=(14, 7), dpi=100)
fft_list = np.asarray(fft_df['fft'].tolist())
fourier_list = [] 
for num_ in [3, 6, 9, 39, 69, 99]:
    fft_list_m10= np.copy(fft_list); fft_list_m10[num_:-num_]=0
    ll = np.fft.ifft(fft_list_m10)
    ll = ll.astype(np.float64)
    fourier_list.append(ll)
    plt.plot(ll, label='Fourier transform with {} components'.format(num_))
# plt.plot(df['freq'],  label='Real')
plt.xlabel('Days')
plt.ylabel('freq')
plt.title('Figure 3: Goldman Sachs (close) stock prices & Fourier transforms')
plt.legend()
plt.show()
```

```python
import pickle 
print(len(fourier_list[0]))
with open('1_fourier.pkl', 'wb') as f:
    pickle.dump(fourier_list, f)
```
